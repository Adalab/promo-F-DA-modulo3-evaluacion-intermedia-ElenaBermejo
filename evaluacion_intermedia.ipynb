```python
#3.Preparaci√≥n de datos
```

```python

dicc_df={}
for col in df.columns:
    cambio=col.lower()
    dicc_df[col]=cambio
    

df.rename(columns =dicc_df, inplace=True)


```

```python
df.head()
```

```python
lista_columnas =["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]

def cambio_puntos(texto):
    try:
        return float(texto.replace(",",""))
    except:
        return np.nan
for columna in lista_columnas:   
   df[columna]= df[columna].apply(cambio_puntos)
```

```python
df[["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]].dtypes
```

```python
lista_cancion =["track","title","album"]
def minuscula(cancion):
    try:
        return cancion.lower()
    except:
        return np.nan
    
for columna in lista_cancion:   
   df[columna]= df[columna].apply(minuscula)
```

```python
df[["track","title","album"]].head(3)
```

```python
def cambio_s(artista):
    try:
        return artista.replace("$","s")
    except:
        return np.nan

for columna in df["artist"]:   
   df["artist"]= df["artist"].apply(cambio_s)
```

```python
df["artist"].sample(20)
```

```python
lista_columnas_likes =["views", "likes"]

for columna in lista_columnas_likes:   
   df[columna]= df[columna].astype('int64')
```
